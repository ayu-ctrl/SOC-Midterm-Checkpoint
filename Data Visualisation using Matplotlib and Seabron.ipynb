```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart with Custom Attributes")
plt.bar(categories, values, color='red', edgecolor='black', linewidth=2)
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart with Custom Attributes")
plt.barh(categories, values, color='red', edgecolor='black', linewidth=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart with Custom Attributes")
sns.barplot(x=categories, y=values, palette= "coolwarm")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

categories = [10, 15, 20, 25]
values = [10, 15, 20, 25]

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Line Chart using Seaborn")

sns.lineplot(x=categories, y=values, marker='o', linestyle='-', color='green')
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [9, 11, 13, 18, 21, 24, 32, 34, 45, 56]

Bins = [0, 10, 20, 30, 40, 50, 60]

plt.xlabel('Value')

plt.ylabel('Frequency')

plt.title('Histogram Example')

plt.hist(data, Bins, histtype='bar', rwidth=0.8, color='cyan')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

categories = [2, 3, 4, 5, 7, 6, 12, 15, 17]

plt.figure(figsize=(4, 4))
sns.histplot(categories, bins=5, color="skyblue")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Histogram using Seaborn")
plt.show()

```
